```python
#Frances Talledo, 6403015, Assignment #1

import pandas as pd

# Data URL
url = "https://raw.githubusercontent.com/AlR0d/courses/main/4421/wage.csv"

# Load the dataset
data = pd.read_csv(url)

# 1. Find the average education level and the lowest and highest years of education
average_education = data['educ'].mean()
min_education = data['educ'].min()
max_education = data['educ'].max()

print(f"Average Education Level: {average_education:.2f}")
print(f"Lowest Years of Education: {min_education}")
print(f"Highest Years of Education: {max_education}")

# 2. Find the average hourly wage
average_wage = data['wage'].mean()
print(f"Average Hourly Wage (in 1976 dollars): ${average_wage:.2f}")

# 3. Find the CPI values for 1976 and 2013
cpi_1976 = 55.6  
cpi_2013 = 233.0  

# 4. Adjust the average hourly wage to 2013 dollars
average_wage_2013_dollars = average_wage * (cpi_2013 / cpi_1976)
print(f"Average Hourly Wage (in 2013 dollars): ${average_wage_2013_dollars:.2f}")

# 5. Count the number of women and men in the sample
num_women = data['female'].sum()  # female column is 1 for women, 0 for men
num_men = data.shape[0] - num_women

print(f"Number of Women in the Sample: {num_women}")
print(f"Number of Men in the Sample: {num_men}")
```

```python
import pandas as pd
import numpy as np

# Load the dataset
data_url = "https://raw.githubusercontent.com/AlR0d/courses/main/4421/meap.csv"
data = pd.read_csv(data_url)

# 1. Find the largest and smallest values of math4. Does the range make sense?
largest_math4 = data['math4'].max()
smallest_math4 = data['math4'].min()
range_math4 = largest_math4 - smallest_math4

print(f"Largest value of math4: {largest_math4}")
print(f"Smallest value of math4: {smallest_math4}")
print(f"Range of math4: {range_math4}")

# 2. Number of schools with a perfect pass rate (100%) on math4 and percentage of the total sample
perfect_pass_math4 = (data['math4'] == 100).sum()
total_schools = data.shape[0]
percentage_perfect_pass_math4 = (perfect_pass_math4 / total_schools) * 100

print(f"Number of schools with perfect pass rate on math4: {perfect_pass_math4}")
print(f"Percentage of schools with perfect pass rate on math4: {percentage_perfect_pass_math4:.2f}%")

# 3. Number of schools with exactly 50% pass rate on math4
exactly_50_math4 = (data['math4'] == 50).sum()
print(f"Number of schools with exactly 50% pass rate on math4: {exactly_50_math4}")

# 4. Compare average pass rates for math4 and read4
average_math4 = data['math4'].mean()
average_read4 = data['read4'].mean()

print(f"Average pass rate for math4: {average_math4:.2f}%")
print(f"Average pass rate for read4: {average_read4:.2f}%")

# Determine which test is harder to pass
if average_math4 < average_read4:
    print("Math test is harder to pass.")
else:
    print("Reading test is harder to pass.")

# 5. Find the correlation between math4 and read4
correlation = data['math4'].corr(data['read4'])
print(f"Correlation between math4 and read4: {correlation:.2f}")

# 6. Average and standard deviation of exppp (expenditure per pupil)
average_exppp = data['exppp'].mean()
std_dev_exppp = data['exppp'].std()

print(f"Average expenditure per pupil (exppp): ${average_exppp:.2f}")
print(f"Standard deviation of expenditure per pupil: ${std_dev_exppp:.2f}")

# Determine if there is wide variation in per pupil spending
if std_dev_exppp > average_exppp * 0.5:
    print("There is wide variation in per pupil spending.")
else:
    print("Variation in per pupil spending is not very wide.")

# 7. Compare percentage difference between two schools
spending_school_A = 6000
spending_school_B = 5500

percentage_difference = 100 * (spending_school_A - spending_school_B) / spending_school_B
log_difference = 100 * (np.log(spending_school_A) - np.log(spending_school_B))

print(f"Percentage difference in spending between School A and School B: {percentage_difference:.2f}%")
print(f"Logarithmic approximation of percentage difference: {log_difference:.2f}%")


```

```python
import pandas as pd

# Load the dataset
data_url = "https://raw.githubusercontent.com/AlR0d/courses/main/4421/countymurders.csv"
data = pd.read_csv(data_url)

# Filter the data for the year 1996
data_1996 = data[data['year'] == 1996]

# 1. Number of counties in the dataset and how many have zero murders
total_counties = data_1996.shape[0]
counties_zero_murders = (data_1996['murders'] == 0).sum()

print(f"Total number of counties: {total_counties}")
print(f"Number of counties with zero murders: {counties_zero_murders}")

# 2. Percentage of counties with zero executions
counties_zero_executions = (data_1996['execs'] == 0).sum()
percentage_zero_executions = (counties_zero_executions / total_counties) * 100

print(f"Percentage of counties with zero executions: {percentage_zero_executions:.2f}%")

# 3. Largest number of murders and executions
largest_murders = data_1996['murders'].max()
largest_executions = data_1996['execs'].max()

print(f"Largest number of murders: {largest_murders}")
print(f"Largest number of executions: {largest_executions}")

# 4. Compute the average number of executions
average_executions = data_1996['execs'].mean()

print(f"Average number of executions: {average_executions:.2f}")

# 5. Compute the correlation coefficient between murders and executions
correlation = data_1996['murders'].corr(data_1996['execs'])

print(f"Correlation coefficient between murders and executions: {correlation:.2f}")

# 6. Discuss the correlation
print("\nAnalysis:")
print("1. The correlation coefficient indicates the strength and direction of the linear relationship between murders and executions. A positive correlation suggests that higher murder counts might be associated with higher execution counts, though correlation does not imply causation.")
print("2. The average number of executions is small, which is expected because executions are relatively rare compared to the number of murders.")
print("3. A positive correlation does not mean that more executions cause more murders. It might reflect that areas with higher crime rates, including more murders, might also have higher execution rates due to a variety of socio-economic and legal factors.")
```

```python
import pandas as pd

# Load the dataset
data_url = "https://raw.githubusercontent.com/AlR0d/courses/main/4421/alcohol.csv"
data = pd.read_csv(data_url)


# 1. Calculate the percentage of men who report abusing alcohol
percentage_abusing_alcohol = (data['abuse'].mean()) * 100
print(f"Percentage of men who report abusing alcohol: {percentage_abusing_alcohol:.2f}%")

# 2. Determine the overall employment rate
employment_rate = (data['employ'].mean()) * 100
print(f"Overall employment rate: {employment_rate:.2f}%")

# 3. Employment rate among men who abuse alcohol
employment_rate_abuse = (data[data['abuse'] == 1]['employ'].mean()) * 100
print(f"Employment rate among men who abuse alcohol: {employment_rate_abuse:.2f}%")

# 4. Employment rate among men who do not abuse alcohol
employment_rate_no_abuse = (data[data['abuse'] == 0]['employ'].mean()) * 100
print(f"Employment rate among men who do not abuse alcohol: {employment_rate_no_abuse:.2f}%")

# 5. Discuss the differences
print("\nDiscussion:")
print("1. The percentage of men abusing alcohol and the employment rates for both groups show how alcohol abuse correlates with employment status.")
print("2. Comparing the employment rate among men who abuse alcohol versus those who do not, you might observe a difference.")
print("3. If men who abuse alcohol have a lower employment rate than those who do not, it suggests a correlation between alcohol abuse and lower employment rates.")
print("4. However, correlation does not imply causation. Other factors might contribute to this relationship. To conclude causation, more detailed analysis and research would be required to control for other variables and use appropriate statistical methods.")
```
